#### transforming columns using maths 

```python
# power transformation - log transformation 
df['avggift'].hist()
```

```python
def log_transform_clean(x):
    x=np.log(x)
    if np.isfinite(x):
        return x
    else:
        return 0
df['avg_gift_log']=list(map(log_transform_clean,df['avggift']))
```
